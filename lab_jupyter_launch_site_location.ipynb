```python
import folium
import pandas as pd
from js import fetch
import io

# Load the data
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Process the data
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Create the map
nasa_coordinate = [29.559684888503615, -95.0830971930759]  # Make sure this is defined
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Define colors
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray']

# Add circles and markers for each unique launch site
for i, site in enumerate(spacex_df['Launch Site'].unique()):
    site_info = spacex_df[spacex_df['Launch Site'] == site].iloc[0]
    color = colors[i % len(colors)]
    
    # Add circle
    folium.Circle(
        location=[site_info['Lat'], site_info['Long']],
        radius=1000,
        color=color,
        fill=True,
        fillColor=color,
        fillOpacity=0.4,
        popup=site
    ).add_to(site_map)
    
    # Add marker
    folium.map.Marker(
        [site_info['Lat'], site_info['Long']],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{site}</b></div>'
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)
#site_map.add_child(lines)
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Assuming spacex_df is already loaded with the data

# Function to calculate distance between two points on Earth
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    return R * c

# Updated coastline point
coastline_point = [28.56217, -80.56777]

# Calculate distances and add to dataframe
spacex_df['Distance_to_Coast'] = spacex_df.apply(lambda row: haversine_distance(row['Lat'], row['Long'], coastline_point[0], coastline_point[1]), axis=1)

# Sort by distance and get the 2 closest sites
closest_sites = spacex_df.sort_values('Distance_to_Coast').drop_duplicates('Launch Site').head(2)

# Create the map
site_map = folium.Map(location=[39.50, -98.35], zoom_start=5)

# Add markers for each launch
marker_cluster = MarkerCluster().add_to(site_map)

for _, launch in spacex_df.iterrows():
    color = 'green' if launch['class'] == 1 else 'red'
    folium.Marker(
        location=[launch['Lat'], launch['Long']],
        popup=f"Launch Site: {launch['Launch Site']}<br>Outcome: {'Success' if launch['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(marker_cluster)

# Add coastline point
folium.Marker(
    location=coastline_point,
    popup="Closest Coastline Point",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(site_map)

# Add PolyLines from closest sites to coastline (now bright blue)
for _, site in closest_sites.iterrows():
    folium.PolyLine(
        locations=[[site['Lat'], site['Long']], coastline_point],
        color='#00BFFF',  # Bright blue (Deep Sky Blue)
        weight=2,
        opacity=0.8
    ).add_to(site_map)

# Fit the map to show all markers
site_map.fit_bounds(site_map.get_bounds())

# Display the map
site_map
```

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Assuming spacex_df is already loaded with the data

# Function to calculate distance between two points on Earth
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    return R * c

# Coastline point
coastline_point = [28.56217, -80.56777]

# Correct coordinates for the closest city
closest_city = [28.61203, -80.80804]

# Calculate distances to coastline and add to dataframe
spacex_df['Distance_to_Coast'] = spacex_df.apply(lambda row: haversine_distance(row['Lat'], row['Long'], coastline_point[0], coastline_point[1]), axis=1)

# Sort by distance and get the 2 closest sites to coastline
closest_sites_coast = spacex_df.sort_values('Distance_to_Coast').drop_duplicates('Launch Site').head(2)

# Calculate distances to closest city and find the closest launch site
spacex_df['Distance_to_City'] = spacex_df.apply(lambda row: haversine_distance(row['Lat'], row['Long'], closest_city[0], closest_city[1]), axis=1)
closest_site_city = spacex_df.sort_values('Distance_to_City').drop_duplicates('Launch Site').iloc[0]

# Create the map
site_map = folium.Map(location=[39.50, -98.35], zoom_start=5)

# Add markers for each launch
marker_cluster = MarkerCluster().add_to(site_map)

for _, launch in spacex_df.iterrows():
    color = 'green' if launch['class'] == 1 else 'red'
    folium.Marker(
        location=[launch['Lat'], launch['Long']],
        popup=f"Launch Site: {launch['Launch Site']}<br>Outcome: {'Success' if launch['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=color, icon='info-sign')
    ).add_to(marker_cluster)

# Add coastline point
folium.Marker(
    location=coastline_point,
    popup="Closest Coastline Point",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(site_map)

# Add closest city point
folium.Marker(
    location=closest_city,
    popup="Closest City",
    icon=folium.Icon(color='purple', icon='info-sign')
).add_to(site_map)

# Add PolyLines from closest sites to coastline
for _, site in closest_sites_coast.iterrows():
    folium.PolyLine(
        locations=[[site['Lat'], site['Long']], coastline_point],
        color='#00BFFF',  # Bright blue (Deep Sky Blue)
        weight=2,
        opacity=0.8
    ).add_to(site_map)

# Add PolyLine from closest site to the closest city
folium.PolyLine(
    locations=[[closest_site_city['Lat'], closest_site_city['Long']], closest_city],
    color='#FF1493',  # Deep Pink
    weight=2,
    opacity=0.8
).add_to(site_map)

# Fit the map to show all markers
site_map.fit_bounds(site_map.get_bounds())

# Display the map
site_map
```
